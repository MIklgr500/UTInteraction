```python
import pandas as pd
```

```python
df1=pd.read_csv('mynet_v4_dr035.csv')
df2=pd.read_csv('mynet_v4_bias.csv')
```

```python
df1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))

ax1.plot(df1['epoch'].values[:200], df1['acc'].values[:200], color='blue', label='train accuracy')
ax1.plot(df1['epoch'].values[:200], df1['val_acc'].values[:200], color='green', label='validation accuracy')
ax1.set_xlabel('epoch')
ax1.set_ylabel('acc')
ax1.set_title('Model with Conv2DLSTM')
ax1.legend(loc='upper center', bbox_to_anchor=(0.7, 0.2),  shadow=True, ncol=1)
ax2.plot(df2['epoch'].values[:], df2['acc'].values[:], color='blue', label='train accuracy')
ax2.plot(df2['epoch'].values[:], df2['val_acc'].values[:], color='green', label='validation accuracy')
ax2.set_xlabel('epoch')
ax2.set_ylabel('acc')
ax2.set_title('Model with DSConv2DLSTM')
ax2.legend(loc='upper center', bbox_to_anchor=(0.7, 0.2),  shadow=True, ncol=1)
fig.savefig('LSTM_network.png')
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,5))

ax.plot(df1['epoch'].values[:150], df1['val_acc'].values[:150], color='blue', label='Conv2DLSTM')
ax.plot(df2['epoch'].values[:150], df2['val_acc'].values[:150], color='green', label='DSConv2DLSTM')
ax.set_xlabel('epoch')
ax.set_ylabel('acc')
ax.legend()

fig.savefig('VrioousDRLSTMConv_v3.png')
```
